```python
%%html
<h1 align="center">ANALYSIS</h1> 
```

```python
 "Top 5 Movies We want According Their Gross"
```

```python
top5movie= df[['movie_title','gross']].head()
top5movie
```

```python
df = df.sort_values(by=['gross'],ascending =False)
df
```

```python
df= df.drop_duplicates(subset=['movie_title'],inplace=False)
```

```python
top10movie= df[['movie_title','gross']].head(10)
top10movie
```

```python
arr= np.arange(len(top10movie.movie_title))
arr
```

```python
"PLOTTING"
```

```python
plt.figure(figsize=(12,5))
plt.xticks(arr,top10movie.movie_title)
plt.xlabel("Movie")
plt.ylabel("Revenue")
plt.title("Top 10 Movies")
plt.bar(arr,top10movie.gross)
plt.xticks(rotation =45)
plt.show()

```

```python
#Top 10 Movies by imdb score
```

```python
df =df.sort_values(by=['imdb_score'],ascending=False)
top10score = df[['movie_title','imdb_score']].head(10)
top10score
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=45)
plt.plot(top10score.movie_title,top10score.imdb_score)
plt.grid()
plt.show()
```
